Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import imutils
from imutils.object_detection import non_max_suppression
```

Read Image

```python
imPath = "img_4.jpg"
im = cv2.imread(imPath)
```

```python
plt.imshow(im[:,:,::-1])
```

Initialize HOG object

```python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
(r, w) = hog.detectMultiScale(im, 
                              winStride=(2, 2), 
                              padding=(16, 16), 
                              scale=0.2)
```

```python
rects = np.array([[x, y, x+w, y+h] for (x, y, w, h) in r])
```

```python
rectsNMS = non_max_suppression(rects, probs=None, overlapThresh=0.6)
```

```python
for (x1, y1, x2, y2) in rectsNMS:
    cv2.rectangle(im, (x1, y1), (x2, y2), (0, 255, 0), 2)

plt.imshow(im[:, :, ::-1])
```
