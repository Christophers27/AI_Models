Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import cvlib
from cvlib.object_detection import draw_bbox
```

Read Image

```python
im = cv2.imread(r'C:\Users\chris\OneDrive\Pictures\Training Images\face_1.jpg')
```

```python
plt.imshow(im[:,:,::-1])
```

Get boundary boxes

```python
bbox, conf = cvlib.detect_face(im)
```

```python
labels = ['person' for i in range(len(conf))]
```

Draw boundary boxes

```python
imLabeled = draw_bbox(im, bbox, labels, conf)
plt.imshow(imLabeled[:,:,::-1])
```
